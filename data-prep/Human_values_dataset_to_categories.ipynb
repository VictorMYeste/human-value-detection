```python
# Upload the labels.tsv file
from google.colab import files
uploaded = files.upload()
```

```python
# Load the dataset
file_name = list(uploaded.keys())[0]
df = pd.read_csv(file_name, sep="\t")
```

```python
# Process the dataset to add new category columns
for category, labels in categories.items():
    # Set new category to 1 if any corresponding old label has a value > 0
    df[category] = (df[labels] > 0).any(axis=1).astype(float)
```

```python
# Calculate the percentage of sentences with a 1 in each category
category_percentages = (df[categories.keys()].sum() / len(df)) * 100

# Display the percentages
for category, percentage in category_percentages.items():
    print(f"{category}: {percentage:.2f}%")
```

```python
# Filter rows where Presence = 1
df_presence = df[df["Presence"] == 1]

# Calculate the percentage of sentences with a 1 in each category for this subset
category_percentages_presence = (df_presence[list(categories.keys())].sum() / len(df_presence)) * 100

# Display the percentages for the subset where Presence = 1
for category, percentage in category_percentages_presence.items():
    print(f"{category}: {percentage:.2f}%")
```

```python
# Save the transformed dataset to a new file
output_file = "transformed_labels.tsv"
df.to_csv(output_file, sep="\t", index=False)

# Download the transformed file
files.download(output_file)
```
