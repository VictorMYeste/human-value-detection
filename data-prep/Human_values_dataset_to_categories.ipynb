```python
# Upload the labels.tsv file
from google.colab import files
uploaded = files.upload()
```

```python
# Load the dataset
file_name = list(uploaded.keys())[0]
df = pd.read_csv(file_name, sep="\t")
```

```python
# Process the dataset to add new category columns
for category, labels in categories.items():
    # Set new category to 1 if any corresponding old label has a value > 0
    df[category] = (df[labels] > 0).any(axis=1).astype(float)
```

```python
# First, collect all labels to remove
labels_to_remove = [
    label for labels in categories.values() for label in labels
]

# Drop duplicates, as labels may appear in multiple categories
labels_to_remove = list(set(labels_to_remove))

# Remove these labels from your DataFrame
df.drop(columns=labels_to_remove, inplace=True, errors='ignore')
```

```python
# Calculate the percentage of sentences with a 1 in each category
category_percentages = (df[categories.keys()].sum() / len(df)) * 100

# Display the percentages
for category, percentage in category_percentages.items():
    print(f"{category}: {percentage:.2f}%")
```

```python
# Filter rows where Presence = 1
df_presence = df[df["Presence"] == 1]

# Calculate the percentage of sentences with a 1 in each category for this subset
category_percentages_presence = (df_presence[list(categories.keys())].sum() / len(df_presence)) * 100

# Display the percentages for the subset where Presence = 1
for category, percentage in category_percentages_presence.items():
    print(f"{category}: {percentage:.2f}%")
```

# Data Augmentation Fix

```python
import pandas as pd
from IPython.display import display

# Load the TSV file into a DataFrame
file_path = 'labels-cat-aug.tsv'  # Update with the correct file path
df = pd.read_csv(file_path, sep='\t')

# List of value columns
value_columns = [
    'Self-direction: thought',
    'Self-direction: action',
    'Stimulation',
    'Hedonism',
    'Achievement',
    'Power: dominance',
    'Power: resources',
    'Face',
    'Security: personal',
    'Security: societal',
    'Tradition',
    'Conformity: rules',
    'Conformity: interpersonal',
    'Humility',
    'Benevolence: caring',
    'Benevolence: dependability',
    'Universalism: concern',
    'Universalism: nature',
    'Universalism: tolerance'
]

# Loop through each value column and merge the "attained" and "constrained" columns
for value in value_columns:
    attained_col = f'{value} attained'
    constrained_col = f'{value} constrained'

    # Create a new column that combines both attained and constrained
    df[value] = (df[attained_col] != 0.0) | (df[constrained_col] != 0.0)

    # Convert boolean values (True/False) to float (1.0/0.0)
    df[value] = df[value].astype(float)

    # Drop the individual attained and constrained columns
    df.drop([attained_col, constrained_col], axis=1, inplace=True)

# Reorder columns: Text-ID, Sentence-ID, then the new unified labels, followed by the remaining columns
new_column_order = ['Text-ID', 'Sentence-ID'] + value_columns + [col for col in df.columns if col not in ['Text-ID', 'Sentence-ID'] + value_columns]

# Apply the new column order
df = df[new_column_order]

# Display the resulting DataFrame
display(df)

# Save the modified DataFrame to a new file (optional)
df.to_csv('labels-cat-aug_new.tsv', sep='\t', index=False)
```
